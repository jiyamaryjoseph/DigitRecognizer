```python
import pandas as pd
from sklearn.datasets import load_digits
digits=load_digits()
df=pd.DataFrame(data=digits.data,columns=digits.feature_names)
df.head()
```

```python
df['Target']=digits.target
df.head()
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(digits.data,digits.target,test_size=.25,random_state=10)
```

```python
model_params={
    'SVC':{
        'model':SVC(gamma='auto'),
        'param':{
            'C':[1,5,8,10],
            'kernel':['linear','rbf']
            }
    },
    'RandomForestClassifier':{
        'model':RandomForestClassifier(),
        'param':{
            'n_estimators':[1,5,10],
            'criterion':["gini", "entropy"],
            'max_depth':[2,3,4],
            'min_samples_split':[2,3,4],
            'min_samples_leaf':[2,3]
            
            }
    },
    'GaussianNB':{
        'model':GaussianNB(),
        'param':{
        }
    },
     'logistic_regression' : {
        'model': LogisticRegression(solver='liblinear',multi_class='auto'),
        'param': {
            'C': [1,5,10]
        }
     },
    'MultinomialNB':{
        'model':MultinomialNB(),
        'param':{
        }
    },
    'decision_tree': {
        'model': DecisionTreeClassifier(),
        'param': {
            'criterion': ['gini','entropy'],
        }
    }
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
```

```python
score=[]
for model_name,mp in model_params.items():
    #clf=RandomizedSearchCV(estimator=mp['model'],param_distributions=mp['param'],cv=5,return_train_score=False,n_iter=100,random_state=10,n_jobs=1)
    clf=GridSearchCV(estimator=mp['model'],param_grid=mp['param'],cv=5,return_train_score=False)
    clf.fit(X_train,y_train)
    score=({
        'model':model_name,
        'best_score':clf.best_score_,
        'best_param':clf.best_params_
    })
df = pd.DataFrame(score,columns=['model','best_score','best_param'])
df
```

```python
model=DecisionTreeClassifier(criterion='entropy')
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
